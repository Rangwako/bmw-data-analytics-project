## Load the data

Load the data from the "/content/BMW sales data (2010-2024) (1).csv" file into a pandas DataFrame.


**Reasoning**:
Import the pandas library, read the CSV file into a DataFrame, and display the first few rows.



```python
import pandas as pd

df = pd.read_csv("/content/BMW sales data (2010-2024) (1).csv")
display(df.head())
```

## Exploratory data analysis (eda)

### Subtask:
Perform initial EDA to understand the data structure, identify missing values, and get basic statistics.


**Reasoning**:
Display the concise summary, descriptive statistics, and missing value counts of the DataFrame `df` to understand its structure and identify missing values.



```python
df.info()
display(df.describe())
display(df.isnull().sum())
```

## Descriptive analysis

### Subtask:
Calculate descriptive statistics for relevant columns, such as measures of central tendency and dispersion.


**Reasoning**:
Calculate descriptive statistics for the specified columns and the frequency distributions for the categorical columns as per the instructions.



```python
display(df['Price_USD'].describe())
display(df['Sales_Volume'].describe())
display(df['Mileage_KM'].describe())
display(df['Region'].value_counts())
display(df['Fuel_Type'].value_counts())
display(df['Sales_Classification'].value_counts())
```

## Data visualization (eda & descriptive)

### Subtask:
Create visualizations to support the EDA and descriptive analysis, such as histograms, box plots, and scatter plots.


**Reasoning**:
Generate the requested visualizations for EDA and descriptive analysis.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Histogram for Price_USD
plt.figure(figsize=(10, 6))
sns.histplot(df['Price_USD'], kde=True)
plt.title('Distribution of Price_USD')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.show()

# 2. Histogram for Sales_Volume
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales_Volume'], kde=True)
plt.title('Distribution of Sales_Volume')
plt.xlabel('Sales Volume')
plt.ylabel('Frequency')
plt.show()

# 3. Box plot for Price_USD by Sales_Classification
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sales_Classification', y='Price_USD', data=df)
plt.title('Distribution of Price_USD by Sales Classification')
plt.xlabel('Sales Classification')
plt.ylabel('Price (USD)')
plt.show()

# 4. Box plot for Sales_Volume by Sales_Classification
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sales_Classification', y='Sales_Volume', data=df)
plt.title('Distribution of Sales_Volume by Sales Classification')
plt.xlabel('Sales Classification')
plt.ylabel('Sales Volume')
plt.show()

# 5. Scatter plot for Mileage_KM and Price_USD
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Mileage_KM', y='Price_USD', data=df)
plt.title('Relationship between Mileage and Price')
plt.xlabel('Mileage (KM)')
plt.ylabel('Price (USD)')
plt.show()

# 6. Bar plot for counts of each Region
plt.figure(figsize=(12, 6))
sns.countplot(y='Region', data=df, order = df['Region'].value_counts().index)
plt.title('Count of Cars by Region')
plt.xlabel('Count')
plt.ylabel('Region')
plt.show()

# 7. Bar plot for counts of each Fuel_Type
plt.figure(figsize=(10, 6))
sns.countplot(x='Fuel_Type', data=df, order = df['Fuel_Type'].value_counts().index)
plt.title('Count of Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.show()
```

## Prescriptive analysis

### Subtask:
Develop and implement a method for prescriptive analysis. This could involve forecasting, optimization, or recommendation systems depending on the data and the goal.


**Reasoning**:
Calculate the average sales volume for each combination of 'Region' and 'Fuel_Type' and store the results in a new DataFrame for further analysis.



```python
sales_by_region_fuel = df.groupby(['Region', 'Fuel_Type'])['Sales_Volume'].mean().reset_index()
display(sales_by_region_fuel)
```

**Reasoning**:
Identify the combinations with the lowest and highest average sales volume from the calculated results.



```python
lowest_sales = sales_by_region_fuel.nsmallest(5, 'Sales_Volume')
highest_sales = sales_by_region_fuel.nlargest(5, 'Sales_Volume')

print("Combinations with the Lowest Average Sales Volume:")
display(lowest_sales)

print("\nCombinations with the Highest Average Sales Volume:")
display(highest_sales)
```

## Data visualization (prescriptive)

### Subtask:
Visualize the results of the prescriptive analysis to communicate findings and recommendations.


**Reasoning**:
Create a bar plot to visualize the average sales volume for each Region and Fuel Type combination, and highlight the lowest and highest sales volume combinations.



```python
plt.figure(figsize=(14, 8))
barplot = sns.barplot(x='Region', y='Sales_Volume', hue='Fuel_Type', data=sales_by_region_fuel, palette='viridis')
plt.title('Average Sales Volume by Region and Fuel Type')
plt.xlabel('Region')
plt.ylabel('Average Sales Volume')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Fuel Type')
plt.tight_layout()

# Highlight lowest and highest sales combinations (optional, can be done manually or with annotations if needed)
# For simplicity in this plot, we are showing the overall distribution.
# Specific highlights can be added as annotations if required.

plt.show()
```
